EXTRAER INFORMACION

```python
from pandas import DataFrame
import pandas as pd
vacunacion = pd.read_csv("country_vaccinations.csv")
vacunacion = pd.DataFrame(vacunacion)
vacunacion
```

Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64. 

```python
vacunacion['date'] = pd.to_datetime(vacunacion['date'])
vacunacion.info()
```

Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines). 

```python
vacunacion[['total_vaccinations', 'vaccines']].groupby('vaccines').sum()
```

Obtener la cantidad de vacunas aplicadas en todo el mundo. 

```python
vacunacion['total_vaccinations'].sum()
```

Calcular el promedio de vacunas aplicadas por país. 

```python
vacunacion[['total_vaccinations', 'country']].groupby('country').mean()
```

Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo. 

```python

vacunasfecha = vacunacion.query("date == '2021-01-29'")
vacunasfecha2 =vacunasfecha['total_vaccinations'].sum()
vacunasfecha2
```

Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  

```python
coDiferencias = vacunacion
coDiferencias['diferencias'] = coDiferencias['daily_vaccinations'] - coDiferencias['daily_vaccinations_raw']
coDiferencias
```

Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
coDiferencias['date'].max() - coDiferencias['date'].min()
```

 Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter /). 

```python
coCantidad = vacunacion
coCantidad['canVac'] = str(coCantidad['vaccines'])
coCantidad['canVac'] = coCantidad['vaccines'].str.split('/').str.len()
coCantidad
```

Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020. 


```python
antes20 = vacunacion
antes20 = antes20.query("date < '2020-12-20'")
antes20
```

Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado
la vacuna Pfizer.

```python
Pfizer = vacunacion
[Pfizer.vaccines.str.contains('Pfizer')]
Pfizer.head()
```
